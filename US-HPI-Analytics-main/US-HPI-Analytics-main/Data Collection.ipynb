```python
import pandas as pd
import numpy as np
from datetime import datetime
```

# Data Collection

```python
#$$$$$$$$$$$$$Target$$$$$$$$$$$
#HPI
hpi = pd.read_csv('CSUSHPISA.csv') #link given

########Demand#########
#Mortgage Avg Interest
mort = pd.read_csv('MORTGAGE30US.csv') #https://fred.stlouisfed.org/series/MORTGAGE30US

#GDP (economy)
gdp = pd.read_csv('GDP (1).csv') #https://fred.stlouisfed.org/series/GDP 

#Unemployment rate (considering employment)
unemployment = pd.read_csv('UNRATE.csv')  #https://fred.stlouisfed.org/series/UNRATE

#property prices (comes underaffordablity)
price = pd.read_csv('QUSR628BIS.csv')#https://fred.stlouisfed.org/series/QUSR628BIS

#personal income (Comes under affordablity)
pi = pd.read_csv('PI.csv') #https://fred.stlouisfed.org/series/PI - in billion

#population
pop = pd.read_csv('POPTHM.csv')

#immigrants yearwise
immi = pd.read_excel('immigrants.xlsx')#https://www.dhs.gov/immigration-statistics/yearbook/2019/table1

######### Supply ########

#marital status (people evacuating their cuurent houses when married/ divorced, both supply and demand)
marital = pd.read_csv("Table MS-1 (ms1)_Table MS-1.csv")  #https://www.census.gov/data/tables/time-series/demo/families/marital.html

#active listing (the houses listed/avaialble for buyers )
listing = pd.read_csv('ACTLISCOUUS.csv') 

#permit ()
permit = pd.read_csv('PERMIT.csv')#https://fred.stlouisfed.org/series/PERMIT, New Privately-Owned Housing Units Authorized in Permit-Issuing Places: Total Units (PERMIT)

#new home (news houses available either )
newhouse = pd.read_csv('HOUST.csv')#https://fred.stlouisfed.org/series/HOUST
#estimates of housing starts include units in structures being totally rebuilt on an existing foundation

#delinquency rate
delin = pd.read_csv('DRSFRMACBS.csv')#https://fred.stlouisfed.org/series/DRSFRMACBS

l = [hpi,mort,gdp,unemployment,price,pi,pop,marital,immi,listing,permit,newhouse]
```

```python
####Join tables
df1 = pd.merge(hpi,gdp,how = 'left')
df2 = pd.merge(df1,unemployment,how = 'left')
df3 = pd.merge(df2,price,how = 'left')
df4 = pd.merge(df3,pi,how = 'left')
df5 = pd.merge(df4,pop,how = 'left')
df6 = pd.merge(df5,listing,how = 'left')
df7 = pd.merge(df6,permit,how = 'left')
df8 = pd.merge(df7,newhouse,how = 'left')
df8 = pd.merge(df8,delin,how = 'left')
df8
```

```python
#marital processing
for i in range(0,29):
    marital['Year'].iloc[i] = 2021 - i
marital['Year'].iloc[29] = 1990
```

```python
marital = marital.rename(columns={'Year' : 'YEAR'})
```

```python
 marital = marital[['YEAR','Men Married1','Men Unmarried Divorced','Men Unmarried Widowed','Women Married1','Women Unmarried Divorced','Women Unmarried Widowed']]
```

```python
#2 features left
df8['YEAR'] = df8['DATE'].apply(lambda x : x[0:4])
df8
```

```python
df8['YEAR'] = df8['YEAR'].astype(str)
marital['YEAR'] = marital['YEAR'].astype(str)
```

```python
df9 = pd.merge(df8,marital,how = 'left')
df9
```

```python
immi = immi.rename(columns = {'Year' : "YEAR"})
```

```python
immi['YEAR'] = immi['YEAR'].astype(str)
```

```python

final_df = pd.merge(df9,immi,how = 'left',on = 'YEAR')
```
